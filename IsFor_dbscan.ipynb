# Isolation Forest

```python
df_mod1 = pd.read_csv("estandarizados/std_140_1.csv")
model1 = IsolationForest(contamination=0.01,random_state=42)
model1.fit(df_mod1)
df_mod1['anomaly_score'] = model1.decision_function(df_mod1)
df_mod1['anomaly'] = model1.predict(df_mod1.drop(columns=['anomaly_score']))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Asumiendo que ya aplicaste la lógica anterior y tienes 0 (normal) y 1 (anomalía)

# 1. Muestreo seguro: Usamos min() para evitar error si el DF tiene menos de 5000 filas
n_samples = min(5000, len(df_std_140_1))
df_vis = df_std_140_1.sample(n=n_samples, random_state=42)

# 2. CORRECCIÓN DE ORDEN:
# Queremos pintar primero los normales (0) y ENCIMA las anomalías (1).
# Al ordenar ascendente (True), los 0 quedan al principio y los 1 al final (se pintan último).
df_vis = df_vis.sort_values('anomaly', ascending=True)

# 3. CORRECCIÓN DE COLORES:
# En el paso anterior definimos Anomalía=1 y Normal=0.
# Debemos actualizar las llaves del diccionario para que coincidan.
colores = {0: 'royalblue', 1: 'red'}

# 4. Generar el Pairplot
sns.pairplot(
    df_vis,
    hue='anomaly',       
    palette=colores,     
    corner=True,         
    plot_kws={'alpha': 0.6, 's': 15}
)

plt.suptitle("Visualización de Anomalías (Muestra)", y=1.02)
plt.show()
print("")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import math

# 1. Seleccionamos solo las columnas numéricas para graficar (excluyendo la etiqueta)
cols_to_plot = [c for c in df_std_140_1.columns if c != 'anomaly']
n_cols = 3  # Número de gráficos por fila
n_rows = math.ceil(len(cols_to_plot) / n_cols)

# 2. Configuración de la figura
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4 * n_rows))
axes = axes.flatten() # Aplanamos para iterar fácilmente

# Colores (0=Normal=Azul, 1=Anomalía=Rojo)
colores = {0: 'royalblue', 1: 'red'}

# 3. Bucle para crear un histograma por cada variable
for i, col in enumerate(cols_to_plot):
    sns.histplot(
        data=df_std_140_1, 
        x=col, 
        hue='anomaly', 
        palette=colores,
        kde=True,              # Agrega la línea de densidad suave
        stat="density",        # Normaliza para comparar distribuciones, no conteos absolutos
        common_norm=False,     # ¡CLAVE! Escala cada grupo independientemente
        element="step",        # Estilo visual más limpio
        ax=axes[i]
    )
    axes[i].set_title(f'Distribución: {col}')

# 4. Eliminar ejes vacíos si sobran espacios en la grilla
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
