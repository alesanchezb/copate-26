### Vibecode para verificar la distribucion de los valores:

```python
import pandas as pd
import numpy as np

def regla_empirica_pandas(df, columna):
    """
    Verifica la regla empÃ­rica (68-95-99.7) en una columna de un DataFrame.
    """
    # Calcular estadÃ­sticas bÃ¡sicas
    mu = df[columna].mean()
    sigma = df[columna].std()
    total = len(df)
    
    # 2. Definir los lÃ­mites
    limites = {
        '1_sigma': (mu - sigma, mu + sigma),
        '2_sigma': (mu - 2*sigma, mu + 2*sigma),
        '3_sigma': (mu - 3*sigma, mu + 3*sigma)
    }
    
    print(f"--- AnÃ¡lisis de '{columna}' ---")
    print(f"Media: {mu:.2f} | Std: {sigma:.2f}")

    # 3. Calcular porcentajes usando .between()
    # .between() incluye los lÃ­mites por defecto (inclusive=True)
    dentro_1 = df[columna].between(limites['1_sigma'][0], limites['1_sigma'][1]).sum()
    dentro_2 = df[columna].between(limites['2_sigma'][0], limites['2_sigma'][1]).sum()
    dentro_3 = df[columna].between(limites['3_sigma'][0], limites['3_sigma'][1]).sum()

    # 4. Imprimir resultados
    print(f"1Ïƒ ({limites['1_sigma'][0]:.2f} a {limites['1_sigma'][1]:.2f}): "
          f"{(dentro_1/total)*100:.2f}% (Ideal: 68%)")
          
    print(f"2Ïƒ ({limites['2_sigma'][0]:.2f} a {limites['2_sigma'][1]:.2f}): "
          f"{(dentro_2/total)*100:.2f}% (Ideal: 95%)")
          
    print(f"3Ïƒ ({limites['3_sigma'][0]:.2f} a {limites['3_sigma'][1]:.2f}): "
          f"{(dentro_3/total)*100:.2f}% (Ideal: 99.7%)")
```

```python
dfs = {
    "df_140_1": df_140_1,
    "df_140_2": df_140_2,
    "df_145_1": df_145_1,
    "df_145_2": df_145_2,
    "df_150_1": df_150_1,
    "df_150_2": df_150_2,
    "df_155_1": df_155_1,
    "df_155_2": df_155_2,
}

for nombre, df in dfs.items():
    print(f"\n>>> Analizando: {nombre}")
    for col in df.columns:
        regla_empirica_pandas(df, col)
```
