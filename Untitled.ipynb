```python
scaler = StandardScaler()
```

```python
Y_scaled = scaler.fit_transform(y)
```

```python
pca = PCA(n_components=None)
pca.fit(Y_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
accumulate_variance = np.cumsum(explained_variance)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(accumulate_variance) + 1), accumulate_variance, marker='o', linestyle='--')
plt.title('Método del Codo para PCA (Varianza Explicada Acumulada)')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada Acumulada')
plt.grid(True)
plt.axhline(y=0.95, color='r', linestyle='-') # Línea de referencia común (95% de varianza)
plt.text(1, 0.96, '95% umbral', color = 'red', fontsize=10)
plt.show()
```

```python
n_componentes_optimos = np.argmax(accumulate_variance >= 0.95) + 1
print(f"El número óptimo de componentes para el 95% de varianza es: {n_componentes_optimos}")
```
